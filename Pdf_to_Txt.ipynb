### 转endnote下载的pdf

```python
import pandas as pd
import textract
import os

# D:/UIUC/2021SPRING/IS567Text Mining/Assignment1/IS557finalProj
# path = 'D:/EndNote Lib.Data/PDF'
path = 'D:/UIUC/2021SPRING/IS567Text Mining/Assignment1/Test'
subfolder = os.listdir(path)
```

```python
text = []
folder = []
fail = []
i = 0

for j in subfolder:
    try:
        aaa = path + '/' + j
        result = textract.process(aaa + '/' + os.listdir(aaa)[0])
        text.append(result)
        folder.append(j)
        i+=1
        print(i)
    except Exception:     
        pass
        fail.append(j) 
    continue
```

```python
endnote_pdf = pd.DataFrame({'folder': folder, 'text': text})
```

```python
pd.set_option('display.max_colwidth', None)
endnote_pdf
```

```python
endnote_fail = pd.DataFrame({'endnote':fail}) 
```

```python
endnote_fail
```

```python
endnote_pdf.to_csv('D:/UIUC/2021SPRING/IS567Text Mining/Assignment1/IS557finalProj/endnote_pdf.csv.gz')
```

```python
endnote_fail.to_csv('D:/UIUC/2021SPRING/IS567Text Mining/Assignment1/IS557finalProj/endnote_fail.csv')
```
