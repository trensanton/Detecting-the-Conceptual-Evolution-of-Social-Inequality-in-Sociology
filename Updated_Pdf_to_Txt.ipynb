### 转endnote下载的pdf

```python
import pandas as pd
import textract
import os


path = 'D:/EndNote Lib.Data/PDF'
subfolder = os.listdir(path)
```

```python
text = []
folder = []
fail = []
title = []
i = 0

for j in subfolder:
    try:
        aaa = path + '/' + j
        result = textract.process(aaa + '/' + os.listdir(aaa)[0])
        text.append(result)
        folder.append(j)
        title.append(os.listdir(aaa)[0])
        i+=1
        print(i)
    except Exception:     
        pass
        fail.append(j) 
    continue
```

```python
endnote_pdf = pd.DataFrame({'folder': folder, 'text': text, 'title':title})
endnote_pdf['text'] = endnote_pdf['text'].apply(lambda x: x.decode('latin-1'))
```

```python
endnote_fail = pd.DataFrame({'endnote':fail}) 
```

```python
endnote_fail
```

```python
endnote_pdf.to_csv('D:/UIUC/2021SPRING/IS567Text Mining/Assignment1/IS557finalProj/endnote_pdf.csv.gz')
```

```python
endnote_fail.to_csv('D:/UIUC/2021SPRING/IS567Text Mining/Assignment1/IS557finalProj/endnote_fail.csv')
```
